```
%sql
with cte as 
(
select c.name as country , 
       avg(ca.duration) as country_average
from country c
inner join person p 
on c.country_code = substring(p.phone_number,1,3) 
inner join calls ca 
on ca.caller_id = p.id  or ca.callee_id = p.id 
group by c.name
)
select country from cte where country_average > (select avg(duration) from calls)
```
