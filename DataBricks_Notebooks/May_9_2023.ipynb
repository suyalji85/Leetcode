```
df_emp_df=spark.read.option("nullValue","null").csv("dbfs:/FileStore/shared_uploads/suyalji85@gmail.com/emp.csv",header=True,inferSchema=True)
display(df_emp_df)
```

```
from pyspark.sql.functions import to_date 
df_emp_df2 = df_emp_df.withColumn("HIREDATE",to_date("HIREDATE",'yyyy-mm-dd')).fillna({"HIREDATE":'9999-12-31'})
display(df_emp_df2)
```

```
# Creating two YEAR and MONTH new columns based on Hiredate date field 
from pyspark.sql.functions import date_format
df_emp_df2 = df_emp_df2.withColumn("YEAR",date_format("HIREDATE",'yyyy')).withColumn("MONTH",date_format("HIREDATE",'MM')).withColumn("DAY",date_format("HIREDATE",'dd'))
display(df_emp_df2)
```

```
# Using PartitionBY with columns creating Partitions 
df_emp_df2.write.format("delta").partitionBy("YEAR","MONTH").mode("overwrite").saveAsTable("emp_part")
```
