# Calculate a field based on filename

```
from pyspark.sql.functions import input_file_name,col,when
df = spark.read.csv("/scenerios/multiplefiles/",header=True,inferSchema=True)
df = df.withColumn("filename",input_file_name())

df = df.withColumn("code",when (col("filename").contains("uk"),101)
                         .when (col("filename").contains("us"),102)
                         .when (col("filename").contains("india"),500)
                   .otherwise(999)) 

df = df.drop("filename")
                        
display(df)
```
